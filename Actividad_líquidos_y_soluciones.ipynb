La polarizabilidad se define como la facilidad con la que una molécula puede cambiar de forma en respuesta an un campo eléctrico externo. En otras palabras, muestra la capacidad de una molécula para alterar su distribución de carga electrónica cuando se somete an una influencia eléctrica.

Las moléculas con electrones muy móviles y nubes electrónicas extensas y dispersas tienden a ser muy polarizables. Los átomos más grandes, como los que se encuentran en los grupos inferiores del sistema periódico, tienen una mayor polarización que los átomos más pequeños.

La relación entre la polarizabilidad y las fuerzas internas moleculares radica en que una molécula polarizable tiene la capacidad de inducir momentos dipolares en otras moléculas cercanas.

image.png

La diferencia fundamental entre un momento dipolar temporal y un momento dipolar permanente radica en su naturaleza y duración. Los momentos dipolares temporales surgen de fluctuaciones momentáneas en la distribución electrónica y son transitorios, mientras que los momentos dipolares permanentes son intrínsecos y se mantienen constantes debido a diferencias de electronegatividad entre átomos enlazados en una molécula polar.

image.png

Esto es debido a la tensión superficial del agua permite que se forme una superficie curvada en un recipiente lleno hasta el borde, lo que evita que el agua se derrame inmediatamente. Sin embargo, esta propiedad tiene un límite y si se añade agua demás o si se perturba la superficie, puede derramarse.

image.png

En un líquido, las moléculas están relativamente cerca una de la otra y tienen suficiente energía cinética como para moverse libremente pero manteniendo una cierta proximidad. Esto significa que hay una resistencia significativa a la compresión adicional.

Cuando se intenta comprimir un líquido, Se trata de reducir el espacio entre moléculas. Sin embargo, debido a la proximidad relativamente cercana de las moléculas en un líquido, hay una resistencia considerable a este cambio en la disposición espacial.

image.png

Las moléculas de agua son polares. Esto significa que tienen una distribución desigual de carga eléctrica, con una región ligeramente negativa (en el oxígeno) y regiones ligeramente positivas (en los hidrógenos). Esta polaridad permite que el agua forme enlaces de hidrógeno, lo que resulta en fuertes fuerzas de adhesión a otras sustancias polaras, como muchas superficies sólidas. Mientras que el mercurio al no poseer los puentes de hidrógeno, no tienen esta propiedad de adherirse a otras partículas polares, y por eso tiende a formar esferas con el contacto con otros sólidos.

image.png

Cuando el agua se enfría y se convierte en hielo, las moléculas ocupan más espacio y se expanden en la estructura cristalina. Esto significa que una cantidad dada de agua en estado sólido ocupará más volumen que la misma cantidad de agua en estado líquido.

SOLUCIONES

image.png

En la disolución, las moléculas del soluto se separan y rodean por completo a las moléculas del disolvente. Por ejemplo, al disolver sal (NaCl) en agua, los iones Na⁺ y Cl⁻ se dispersan en el agua, interactuando con las moléculas de H₂O a través de atracciones eléctricas, formando una solución homogénea.

image.png

"Lo semejante disuelve lo semejante" se refiere al principio en la que lassustancias con fuerzas intermoleculares similares tienden a disolverse entre sí. Por ejemplo, compuestos iónicos se disuelven en disolventes polares como el agua (con enlaces covalentes polares), ya que ambas tienen interacciones intermoleculares significativas

image.png

La solvatación es el proceso en el cual solventes rodean y estabilizan un soluto mediante interacciones intermoleculares, como enlaces de hidrógeno o fuerzas dipolo-dipolo, formando una disolución homogénea. En los factores que influyen encontramos a la naturaleza del soluto y del solvente, nos referimos con esto a la polaridad, la carga y la estructura molecular de ambos. Entre otros factores encontramos a la temperatura, presión o la superficie de contacto.

image.png

Naturaleza del soluto y del solvente: La polaridad, la carga y la estructura molecular de ambas sustancias afectan la solubilidad. Sustancias polares tienden a disolverse en solventes polares y viceversa.

Temperatura: En general, a temperaturas más altas, la solubilidad de sólidos en líquidos aumenta, mientras que la solubilidad de gases en líquidos disminuye.

Presión: La solubilidad de gases en líquidos puede aumentar con la presión. Generalmente este factor no es tan notorio en líquidos, pero en algunos casos si se llega a considerar.

Cuando dos líquidos son miscibles significa que tienen la capacidad de mezclarse en cualquier proporción y formar una solución homogénea.

UNIDADES DE CONCENTRACIÓN

image.png

a) El porcentaje de masa es la proporción de la masa de un soluto con respecto a la masa total de la solución, expresada como un porcentaje en masa.

b) La fracción molar es la proporción de la cantidad de sustancia de una sustancia en particular con respecto a la cantidad total de sustancia en la solución. Podría decirse que se mide en moles.

c) La molaridad es la concentración de una solución expresada en moles de soluto por litro de solución. Se mide en mol/L o M.

d) La molalidad es la concentración de una solución expresada en moles de soluto por kilogramo de disolvente. Se mide en mol/kg.

De molaridad a molalidad se necesita conocer la densidad del disolvente, luego aplicamos la fórmula de molalidad, dada por M.m/ρ, y luego para pasar de molaridad a porcentaje en masa se necesita conocer las masas de soluto y del disolvente, luego aplicamos la fórmula de % en masa, y para pasar de % en masa a molaridad se necesita conocer las cantidades de sustancia, para usar la fórmula de molaridad, dada por moles de soluto/litros de solución.

```python
#Para a b y c)
masa_slto_a = 5.50
masa_dslc_a = 78.2
masa_total_solucion_a = masa_slto_a + masa_dslc_a
masa_slto_b = 3.10
masa_dslc_b = 152
masa_total_solucion_b = masa_slto_b + masa_dslc_b
masa_slto_c = 4.5
masa_dslc_c = 29
masa_total_solucion_c = masa_slto_c + masa_dslc_c
p_en_masa_a = ((masa_slto_a)/(masa_total_solucion_a))*100
p_en_masa_b = ((masa_slto_b)/(masa_total_solucion_b))*100
p_en_masa_c = ((masa_slto_c)/(masa_total_solucion_c))*100
p_en_masa_a_aprox = round(p_en_masa_a,2)
p_en_masa_b_aprox = round(p_en_masa_b,2)
p_en_masa_c_aprox = round(p_en_masa_c,2)
print("El porcentaje en masa de a es: ",p_en_masa_a_aprox,"%")
print("El porcentaje en masa de b es: ",p_en_masa_b_aprox,"%")
print("El porcentaje en masa de c es: ",p_en_masa_c_aprox,"%")
```
